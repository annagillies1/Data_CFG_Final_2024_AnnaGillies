1. PANDAS (25 points)

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:
import pandas as pd

df = pd.read_csv('netflix_titles.csv', sep=',')
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
df.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows
df.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.


missing_data = df[df.isnull().any(axis=1)]

# Count the number of rows with missing values
num_rows_with_missing = missing_data.shape[0]

# Display the count of rows with missing values
num_rows_with_missing
```

```python
# showing the resulting dataframe 

df.dropna()
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame. Use any method for this.

# filtering 'type' to 'Movie'
movies_df = df[df['type'] == 'Movie']

# displaying the resulting dataframe
movies_df
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies in the dataset and return the resulting data frame
    
# Calculate the average duration of movies in the dataset
average_duration = movies_df['duration'].str.extract('(\d+)').astype(float).mean()

# Replace missing durations with the calculated average
movies_df['duration'] = movies_df['duration'].fillna(f'{average_duration} min')

# Return the resulting DataFrame
movies_df
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.

duplicates_count = movies_df.duplicated().sum()

duplicates_count

# there are no duplicates
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.

# renaming all of the columns to start with a cap letter
movies_df.columns = movies_df.columns.str.title()

# printing the database 
movies_df
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.

movies_df.to_csv('cleaned_movies_dataset.csv')
```

2. MATPLOTLIB (25 points)

Use the netflix dataset to solve the exercises in this section.

```python
# 1 (10 marks). Create a visualisation (by using the plot that is most appropriate for this) which shows the frequency of movies by year of release.

import matplotlib.pyplot as plt
cleaned_movies_df = pd.read_csv('cleaned_movies_dataset.csv', sep=',')

# creating a histogram of movie release years 

# figsize is controling the dimensions of the plot
plt.figure(figsize=(10,6))
plt.hist(cleaned_movies_df['Release_Year'], bins='auto', color='skyblue', edgecolor ='black')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.title('Frequency of Movies by Year of Release')
plt.grid(True)
plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.

df_exploded = cleaned_movies_df.assign(Listed_In=cleaned_movies_df['Listed_In'].str.split(', ')).explode('Listed_In')

# Calculating the frequency of each category 
genre_count = df_exploded['Listed_In'].value_counts()

# Filtering it to the top five 
top_genres = genre_count.head(5)

# Plotting a pie 
plt.figure(figsize=(8,8))
plt.pie(top_genres, labels=top_genres.index)
plt.title('Top Five Popular Genres')
plt.axis('equal')
plt.show()
```
